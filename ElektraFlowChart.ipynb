```python
import math

def Interpolate(x, y, X_Array, Y_Array, Z_Array):
    x0 = np.searchsorted(X_Array, x)
    y0 = np.searchsorted(Y_Array, y)
    if (x0 == 0) and (x < X_Array[0]):
        z0 = Interpolate(X_Array[0], y, X_Array, Y_Array, Z_Array)
        z1 = Interpolate(X_Array[1], y, X_Array, Y_Array, Z_Array)
        if (X_Array[0] < X_Array[1]):
            z = z0 + (z1 - z0) * (x - X_Array[0]) / (X_Array[1] - X_Array[0])
        else:
            z = z0    
    elif (x0 == len(X_Array)) and (x > X_Array[-1]):
        z0 = Interpolate(X_Array[-2], y, X_Array, Y_Array, Z_Array)
        z1 = Interpolate(X_Array[-1], y, X_Array, Y_Array, Z_Array)
        if (X_Array[-2] < X_Array[-1]):
            z = z1 + (z1 - z0) * (x - X_Array[-1]) / (X_Array[-1] - X_Array[-2])
        else:
            z = z1
    else:
        if (x0 == 0):
            x1 = 0
        elif (x0 == len(X_Array)):
            x0 = len(X_Array) - 1
            x1 = x0
        else:
            if (x == X_Array[x0]):
                x1 = x0
            else:
                x1 = x0
                x0 = x0 - 1
        
        if (y0 == 0):
            y1 = 0
        elif (y0 == len(Y_Array)):
            y0 = len(Y_Array) - 1
            y1 = y0
        else:
            if (y == Y_Array[x0]):
                y1 = y0
            else:
                y1 = y0
                y0 = y0 - 1
        
        if (x0 == x1) and (y0 == y1):
            z = Z_Array[y0,x0]
        elif (x0 == x1):
            z0 = Z_Array[y0,x0]
            z1 = Z_Array[y1,x0]
            z = z0 + (z1 - z0) * (y - Y_Array[y0]) / (Y_Array[y1] - Y_Array[y0])
        elif (y0 == y1):
            z0 = Z_Array[y0,x0]
            z1 = Z_Array[y0,x1]
            z = z0 + (z1 - z0) * (x - X_Array[x0]) / (X_Array[x1] - X_Array[x0])
        else:
            z00 = Z_Array[y0,x0]
            z01 = Z_Array[y0,x1]
            z0 = z00 + (z01 - z00) * (x - X_Array[x0]) / (X_Array[x1] - X_Array[x0])
            z10 = Z_Array[y1,x0]
            z11 = Z_Array[y1,x1]
            z1 = z10 + (z11 - z10) * (x - X_Array[x0]) / (X_Array[x1] - X_Array[x0])
            z = z0 + (z1 - z0) * (y - Y_Array[y0]) / (Y_Array[y1] - Y_Array[y0])
        
    return(z)
            
def GetElektraGasFlow(ElektraType, GasP, GasDP, GasT, GasNormDensity, ThrottlePos):
# ElektraType [0,1,2,3,4,5,6] = [Elektra50, 85, 110, 160, 30, 42, II-110]
# GasP [bar]
# GasDP [bar]
# GasT [°C]
# GasNormDensity [kg/Nm³]
# ThrottlePos [%]
    
    if GasP >= GasDP/2:
        GasPress = GasP - GasDP/2
    else:
        GasPress = 0
        
    GasDensity = GasNormDensity * (GasPress/1.013) * (273/(GasT+273)) # kg/m³
        
    GasVelocity = math.sqrt(2 * 100000 * GasDP / GasDensity) # m/s
    
    ElektraCalibrationValue = Interpolate(GasDP, ThrottlePos, ElektraCalibration_Area_Const_X, ElektraCalibration_Area_Const_Y, ElektraCalibration_Area_Const[ElektraType])
    GMUArea = ElektraCalibrationValue * ElektraMaxArea[ElektraType] / 100 # mm²
    
    if (GasVelocity > 0 and GMUArea > 0):
        NormGasFlow = 3600 * (GMUArea / 1000000) * GasVelocity * (GasDensity / GasNormDensity)
    else:
        NormGasFlow = 0
        
    return(NormGasFlow)       
```
